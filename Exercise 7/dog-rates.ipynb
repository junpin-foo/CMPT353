```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import datetime
from scipy import stats
import seaborn as sns
```

```python

data = pd.read_csv("dog_rates_tweets.csv", parse_dates= ['created_at'])
```

```python
def to_timestamp(data): #2018-05-03 18:29:35 example format
    
    ret = data.timestamp()
    return ret

```

```python
data['timestamp'] = data['created_at'].apply(to_timestamp)
```

```python
def matching(data):
    result = re.search(r'(\d+(\.\d+)?)/10', data)

    if result:
        ret = float(result.group(1))
        if(ret > 25):
            return float("nan")
        else:
            return ret

 
data['score'] = data['text'].apply(matching)
data["NotNa"] = data['score'].notna()
data = data.dropna()
```

```python
reg = stats.linregress(data['timestamp'], data['score'])

sns.set()
plt.figure(facecolor='white')
plt.xticks(rotation=25)
plt.plot(data['created_at'], data['score'], 'b.', alpha=0.5)
plt.plot(data['created_at'], data['timestamp']*reg.slope + reg.intercept, 'r-', linewidth=3)
plt.savefig('dog-rates-result.png')
```

```python
print(reg.pvalue)

sns.set()
residuals = data['score'] - (reg.slope*data['timestamp'] + reg.intercept)
plt.hist(residuals, bins=20)
plt.savefig("dog-rates-residual-hist.png")
```
